# Overview 

This notebook attempts to determine which products are bought together as a **starting package**.

# Method

More precisely, we search for which products are bought **together** in the **first year** of transaction of each organization.

This is done simply by selecting products that were bought in the first year and counting their correlation. 
No fancy math involved but a lot of data wrangling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# Load the data and extract the first year of transaction for each org

```python
# The data file can be obtained from the Google drive
df = pd.read_csv("../Data/techsoup_datakind_details_wide.csv", parse_dates=9)
```

```python
# Just to check that the notebook works, one can restrict the data 
# (faster execution) by uncommenting the following line.
#df = df.head(10000)
```

```python
# Make sure the data is sorted by year
df = df.sort_index(by='transaction_date')
```

```python
# Extract the year. This takes time because of the date parsing !
df['year'] = pd.DatetimeIndex(df.transaction_date).year
```

```python
# Find first year through groupby 
first_year_by_org = df[['org','year']].groupby('org', as_index=False).first()
# Rename variable
first_year_by_org['first_year'] = first_year_by_org['year'].copy()
del first_year_by_org['year']
```

```python
# Associate the first year to the global data
df = pd.merge( df, first_year_by_org[['org', 'first_year']], on='org')
```

# Prepare the data needed to do the correlation matrix

## Retain only the first transaction years

```python
# Retain only a few field
# and retain only the transactions that were made 
# during the first year of transaction for each organization
first_year_orders = df[['org', 'item', 'first_year']][df['year']==df['first_year']]
```

```python
first_year_orders.head(5)
```

## Add new fields for the top N products bought during the first year

```python
# Select the top N products
Ntop = 10

top_products = first_year_orders.item.value_counts().head(Ntop).keys()
top_count = first_year_orders.item.value_counts().head(Ntop).values

plt.title('Top %d products bought in the first year' %Ntop, fontsize=13)
plt.bar( np.arange(Ntop), top_count, width=0.5 )

# Modify labels
plt.xticks(np.arange(Ntop), top_products, rotation='vertical')

plt.ylabel('Overall count (no correlation)')
plt.xlabel('Top %d product' %Ntop)
```

```python
# Create a new field for each of these products
# Fill these fields with a 1 if they were bought 
# and 0 otherwise
# (number of licenses not taken into account)
for product in top_products :
    first_year_orders[product] = 0.
    first_year_orders[product][first_year_orders['item']==product]=1
```

```python
first_year_orders.head(10)
```

## Collapse the above data to have only one line per organization

```python
items_orders_by_org = first_year_orders.groupby('org')[top_products].sum()
```

```python
items_orders_by_org.tail(10)
```

## Find the organizations that bought more than just one product in their first year

```python
# Sum the total number of orders
items_orders_by_org['tot_order'] = items_orders_by_org.sum(axis=1)
```

```python
# Select only the organization
several_orders_by_org = items_orders_by_org[ items_orders_by_org['tot_order'] > 1 ]
```

```python
several_orders_by_org.head(10)
```

# Build the correlation matrix and fill it

```python
# Matrix with one entry for each top product
corr = np.zeros((Ntop,Ntop)) 
```

```python
# Fill the matrix

# Go through the organizations
for i in range(len(several_orders_by_org)) :
    # Loop through the items that this org bought in its first year 
    for i1 in range(Ntop) :
        # Check that the org bought it
        if several_orders_by_org[top_products[i1]].iloc[i] != 0 :
            # Loop through the items again for correlation
            for i2 in range(Ntop) :
                # Check that the org bought it
                if several_orders_by_org[top_products[i2]].iloc[i] != 0 :
                    # Check that it is not the same item
                    if i1 != i2 :
                        corr[i1,i2] += 1
```

# Final plot

```python
plt.figure(figsize=(5,12))

plt.suptitle('Starting package : \n Products bought together by each organization \n in their first year of transaction',
             fontsize=15)

# Correlation matrix
plt.subplot(211)
ax = plt.imshow( corr, interpolation='nearest', cmap='Blues', aspect='auto' )
plt.ylabel('Top %d products' %Ntop)
# Modify labels
plt.xticks(np.arange(Ntop), top_products, rotation='vertical')
plt.yticks(np.arange(Ntop), top_products)

# Global count
plt.subplot(212)
plt.bar( np.arange(Ntop),  top_count )
plt.ylabel('Overall count (no correlation)')
# Modify labels
plt.xticks(np.arange(Ntop), top_products, rotation='vertical')

plt.savefig('Correlation.png')
```
